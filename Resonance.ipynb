```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
resonance_df = pd.read_csv("resonance1.txt", delimiter="	", header=None)
resonance_df.dropna(how='all', axis=1, inplace=True) 
resonance_df.columns = ["Frequency", "Ratio", "InputV", "OutputV"]
print(resonance_df)
```

```python
colors = ['green', 'orange', 'blue']
label = ['res1' ,'23pf', '250pf']

plt.plot(  
        resonance_df["Frequency"],        
        resonance_df["Ratio"],  
        color=colors[0],              
        label=label[0]   
                     )

plt.plot(  
        resonance_df["Frequency"],        
        resonance_df["Ratio"],  
        color=colors[1],              
        label=label[1]   
                     )

plt.scatter(  
        resonance_df["Frequency"],        
        resonance_df["Ratio"],  
        color=colors[2],              
        label=label[2]   
                     )


plt.xlabel('Frequency')
plt.ylabel('Ratio')
plt.legend(loc='lower right')

plt.show()
```
